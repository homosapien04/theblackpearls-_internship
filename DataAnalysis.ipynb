```python
import pandas as pd

# Load the dataset with a specified encoding
df = pd.read_csv("C:/Users/Harshitha/Desktop/black_pearls_finals/sales_data_sample.csv", encoding='latin1')

```

```python
# Filter data where status is 'shipped'
shipped_orders = df[df['STATUS'] == 'shipped']
print(shipped_orders)
```

```python
# Sort data by sales in descending order
sorted_data = df.sort_values(by='SALES', ascending=False)
print(sorted_data)
```

```python
# Group data by year_id and calculate total sales for each year
grouped_data = df.groupby('YEAR_ID')['SALES'].sum().reset_index()
print(grouped_data)
```

```python
# Calculate summary statistics for sales
mean_sales = df['SALES'].mean()
median_sales = df['SALES'].median()
std_sales = df['SALES'].std()

print(f"Mean Sales: {mean_sales}")
print(f"Median Sales: {median_sales}")
print(f"Standard Deviation of Sales: {std_sales}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of sales
plt.figure(figsize=(10, 6))
sns.histplot(df['SALES'], kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot of total sales by year
plt.figure(figsize=(10, 6))
sns.barplot(x='YEAR_ID', y='SALES', data=grouped_data)
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Scatter plot of sales vs. quantity
plt.figure(figsize=(10, 6))
sns.scatterplot(x='QUANTITYORDERED', y='SALES', data=df)
plt.title('Sales vs. Quantity')
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='CITY', y='SALES', data=df)
plt.title('Sales by City')
plt.xlabel('City')
plt.ylabel('Sales')
plt.show()
plt.show()
```
